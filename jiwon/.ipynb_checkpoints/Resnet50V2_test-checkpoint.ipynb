```python
import os
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
import dlib
```

```python
def load_data(data_dir, img_size=(224, 224)):
    """
    Load dataset from directory.

    Args:
    - data_dir (str): Path to the directory containing the dataset.
    - img_size (tuple): Size to resize the images.

    Returns:
    - images (list): List of resized images.
    - labels (list): List of corresponding labels.
    """
    images = []
    labels = []
    for label, category in enumerate(['Closed_Eyes', 'Open_Eyes']):
        category_dir = os.path.join(data_dir, category)
        for img_name in os.listdir(category_dir):
            img_path = os.path.join(category_dir, img_name)
            img = cv2.imread(img_path)
            img = cv2.resize(img, img_size)
            images.append(img)
            labels.append(label)
    return np.array(images), np.array(labels)
```

```python
def prepare_data(data_dir, test_size=0.2, img_size=(224, 224)):
    """
    Prepare dataset for training and testing.

    Args:
    - data_dir (str): Path to the directory containing the dataset.
    - test_size (float): Percentage of the dataset to use for testing.
    - img_size (tuple): Size to resize the images.

    Returns:
    - train_set (tuple): Tuple containing train images and labels.
    - test_set (tuple): Tuple containing test images and labels.
    """
    # Load data
    images, labels = load_data(data_dir, img_size)
    
    # Split into train and test sets
    train_images, test_images, train_labels, test_labels = train_test_split(images, labels, test_size=test_size, random_state=42)
    
    # Normalize images
    train_images = train_images.astype('float32') / 255.0
    test_images = test_images.astype('float32') / 255.0
    
    return (train_images, train_labels), (test_images, test_labels)
```

```python
# Example usage:
(train_images, train_labels), (test_images, test_labels) = prepare_data(r'C:\Users\impra\Desktop\2024_first_semester\AIUI\Stracker\eye-detection-model\jiwon\drowsy_dataset')
```

```python
# Define ResNet-50V2 model
base_model = ResNet50V2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

# Train model
model.fit(train_images, train_labels, batch_size=32, epochs=2, validation_data=(test_images, test_labels))

# Evaluate model
loss, accuracy = model.evaluate(test_images, test_labels)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```

```python
def detect_face_and_eyes(frame):
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)# Convert frame to grayscale
   
    face_detector = dlib.get_frontal_face_detector() # Load face detector
   
    faces = face_detector(gray) # Detect faces
    for face in faces:
        x, y, w, h = face.left(), face.top(), face.width(), face.height()# Draw blue rectangle around face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        eye_detector = dlib.get_frontal_face_detector()# Load eye detector
        eyes = eye_detector(gray[y:y+h, x:x+w])# Detect eyes
        
        for eye in eyes:
            ex, ey, ew, eh = eye.left(), eye.top(), eye.width(), eye.height()
            cv2.rectangle(frame, (ex+x, ey+y), (ex+x+ew, ey+y+eh), (0, 0, 255), 2)# Draw red rectangle around eyes
            
            # Predict if eye is open or closed
            eye_img = frame[ey+y:ey+y+eh, ex+x:ex+x+ew]
            eye_img = cv2.resize(eye_img, (224, 224))
            eye_img = np.expand_dims(eye_img, axis=0)
            eye_img = eye_img.astype('float32') / 255.0
            
            prediction = model.predict(eye_img)
            if prediction > 0.5:
                cv2.putText(frame, 'Open', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
            else:
                cv2.putText(frame, 'Closed', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
                
    return frame
```

```python
# Open webcam
cap = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Function to predict eye state
def predict_eye_state(roi_gray):
    # Resize and preprocess the image
    img = cv2.resize(roi_gray, (224, 224))  # Resize the image to match model input shape
    img = np.expand_dims(img, axis=-1)  # Add channel dimension
    img = np.repeat(img, 3, axis=-1)  # Repeat the image to have 3 channels (for RGB)
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    img = img.astype('float32') / 255.0

    # Predict eye state
    prediction = model.predict(img)
    return prediction[0][0]

# Open webcam
cap = cv2.VideoCapture(0)

# Initialize variables for previous eye state
prev_eye_state = None

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to grayscale for face and eye detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        # Draw rectangle around the face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]

        # Detect eyes within the face
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in eyes:
            # Check if the eye region is within the expected range (not near the nose)
            if ey > h / 5:
                # Draw rectangle around each eye
                cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 0, 255), 2)

                # Predict eye state
                eye_roi_gray = roi_gray[ey:ey+eh, ex:ex+ew]
                prediction = predict_eye_state(eye_roi_gray)

                # Display eye state text above the frame if it has changed
                eye_state = "Open" if prediction > 0.4 else "Closed"
                if eye_state != prev_eye_state:
                    cv2.putText(frame, eye_state, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
                    prev_eye_state = eye_state

    # Show the frame
    cv2.imshow('Eye State Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all windows
cap.release()
cv2.destroyAllWindows()
```
