```python
import cv2
import numpy as np
from keras.models import load_model

# 딥러닝 기반 얼굴 감지 모델 로드
face_net = cv2.dnn.readNetFromCaffe(r"C:\Users\impra\Desktop\2024_first_semester\AIUI\Stracker\eye-detection-model\jiwon\deploy.prototxt", r"C:\Users\impra\Desktop\2024_first_semester\AIUI\Stracker\eye-detection-model\jiwon\res10_300x300_ssd_iter_140000.caffemodel")

# 눈을 감았는지 떳는지 판단하는 모델 로드
eye_model = load_model("Improved_Resnet50v2_model.keras")

# 실시간 웹캠 캡처
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # 이미지 크기 변경 및 전처리
    blob = cv2.dnn.blobFromImage(cv2.resize(frame, (300, 300)), 1.0, (300, 300), (104.0, 177.0, 123.0))

    # 얼굴 감지
    face_net.setInput(blob)
    detections = face_net.forward()

    # 감지된 얼굴에 사각형 그리기 및 눈 감음 감지
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        if confidence > 0.5:  # 얼굴을 감지한 경우
            box = detections[0, 0, i, 3:7] * np.array([frame.shape[1], frame.shape[0], frame.shape[1], frame.shape[0]])
            (startX, startY, endX, endY) = box.astype("int")
            face_roi = frame[startY:endY, startX:endX]

            # 얼굴 영역에서 눈 감음 감지
            eye_roi = cv2.resize(face_roi, (224, 224))
            eye_roi = np.expand_dims(eye_roi, axis=0)  # 배치 차원 추가
            eye_roi = eye_roi / 255.0
            prediction = eye_model.predict(eye_roi)
            if prediction[0][0] > 0.5:  # 양쪽 눈이 감겼을 때
                cv2.putText(frame, 'Eye Closed', (startX, startY-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)
            else:  # 하나 이상의 눈이 떴거나 감지하지 못한 경우
                cv2.putText(frame, 'Eye Open', (startX, startY-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

            # 얼굴 영역에 사각형 그리기
            cv2.rectangle(frame, (startX, startY), (endX, endY), (0, 255, 0), 2)

    # 화면에 출력
    cv2.imshow('Face Detection with Eye Blink Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 해제
cap.release()
cv2.destroyAllWindows()
```
