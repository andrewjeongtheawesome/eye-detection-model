```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.resnet_v2 import preprocess_input
import dlib

# 모델 로드
model = load_model('Resnet50v2_model.keras')

# 얼굴 검출을 위한 dlib 얼굴 검출기 로드
detector = dlib.get_frontal_face_detector()

# 웹캠으로부터 비디오 캡쳐
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # 이미지를 RGB로 변환
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # 얼굴 검출
    faces = detector(rgb_frame)
    
    for face in faces:
        x1, y1, x2, y2 = face.left(), face.top(), face.right(), face.bottom()
        
        # 얼굴 영역 잘라내기
        face_img = frame[y1:y2, x1:x2]
        
        # 이미지 전처리 및 크기 조정
        resized_img = cv2.resize(face_img, (224, 224))
        preprocessed_img = preprocess_input(resized_img)
        preprocessed_img = np.expand_dims(preprocessed_img, axis=0)
        
        # 눈 감음 예측
        prediction = model.predict(preprocessed_img)
        
        # 예측 결과에 따라 텍스트 박스 삽입
        if prediction < 0.45:
            cv2.putText(frame, 'Eyes Closed', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)
        else:
            cv2.putText(frame, 'Eyes Open', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        
        # 얼굴 주위에 사각형 그리기
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
    
    # 프레임 표시
    cv2.imshow('Webcam', frame)
    
    # 종료 키
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 자원 해제
cap.release()
cv2.destroyAllWindows()
```
